```python
import pandas as pd
```

```python
df = pd.read_csv("mood_dataset.csv")
print(df.head())
print(df.columns)
print(df['mood'].value_counts())
```

```python
df = df.rename(columns={'Tweet': 'text'})
df = df[['text', 'mood']].dropna()
df['mood'] = df['mood'].str.capitalize()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['mood'], test_size=0.2, random_state=42)

# Buat pipeline model
model = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1,2))),
    ('clf', LogisticRegression(max_iter=200))
])

# Latih model
model.fit(X_train, y_train)

# Evaluasi
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, "model-mood-classifier.joblib")
```
